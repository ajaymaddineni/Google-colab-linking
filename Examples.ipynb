# Example #1

```python
import numpy as np
import matplotlib.pyplot as plt
A=np.array([1,2,3,4,5])
print(A)
B=2*A^3
C=2*A^2
print(B)
print(C)
plt.plot(A,B, label='F3(x)')
plt.plot(A,C, label='F2(x)')
plt.xlabel('X')
plt.ylabel("F(x)")
plt.legend(loc='lower right')
```
